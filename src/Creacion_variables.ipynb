```python
def goles(season2223_1):
    teams = {}
    for i in season2223_1.groupby('HomeTeam').mean().T.columns:
        teams[i] = []
    #print(teams, len(teams))

    for i in range(len(season2223_1)):
        ATGC = season2223_1.iloc[i]['FTHG']
        HTGC = season2223_1.iloc[i]['FTAG']
        teams[season2223_1.iloc[i].HomeTeam].append(HTGC)
        teams[season2223_1.iloc[i].AwayTeam].append(ATGC)
    #print(teams, len(list(teams.values())[0]))

    GoalsConceded = pd.DataFrame(data=teams, index = [i for i in range(1,len(list(teams.values())[0])+1)]).T
    GoalsConceded[0] = 0

    for i in range(2,len(list(teams.values())[0])+1):
        GoalsConceded[i] = GoalsConceded[i] + GoalsConceded[i-1]
    GoalsConceded
    teams = {}
    for i in season2223_1.groupby('HomeTeam').mean().T.columns:
        teams[i] = []
    #print(teams, len(teams))

    for i in range(len(season2223_1)):
        HTGS = season2223_1.iloc[i]['FTHG']
        ATGS = season2223_1.iloc[i]['FTAG']
        teams[season2223_1.iloc[i].HomeTeam].append(HTGS)
        teams[season2223_1.iloc[i].AwayTeam].append(ATGS)
    #print(teams, len(list(teams.values())[0]))

    GoalsScored = pd.DataFrame(data=teams, index = [i for i in range(1,len(list(teams.values())[0])+1)]).T
    GoalsScored[0] = 0

    for i in range(2,len(list(teams.values())[0])+1):
        GoalsScored[i] = GoalsScored[i] + GoalsScored[i-1]
    GoalsScored
    season2223_1 = season2223_1.sort_values(by='Id',ascending=False)
    j = 0
    HTGS = []
    ATGS = []
    HTGC = []
    ATGC = []
    for i in range(len(season2223_1)):
        ht = season2223_1.iloc[i].HomeTeam
        at = season2223_1.iloc[i].AwayTeam
        HTGS.append(GoalsScored.loc[ht][j])
        ATGS.append(GoalsScored.loc[at][j])
        HTGC.append(GoalsConceded.loc[ht][j])
        ATGC.append(GoalsConceded.loc[at][j])
        
        if ((i + 1)% 11) == 0 and len(teams) == 22:
            j = j + 1
        elif ((i + 1)% 10) == 0 and len(teams) == 20:
            j = j + 1

    season2223_1['HTGS'] = HTGS
    season2223_1['ATGS'] = ATGS
    season2223_1['HTGC'] = HTGC
    season2223_1['ATGC'] = ATGC

    season2223_1 = season2223_1.sort_values(by='Id',ascending=True)
    print('Completada')
    print('Temporada: ' + str(season2223_1.Season.unique()) + ', división: ' + season2223_1.Div.unique())
    return season2223_1
```

```python
def get_points(result):
    if result == 'W':
        return 3
    elif result == 'D':
        return 1
    else:
        return 0

match_point = pd.DataFrame()
teams = {}
def get_mp(sesion):
    global teams
    teams = {}
    for i in sesion.groupby('HomeTeam').mean().T.columns:
        teams[i] = []

    for i in range(len(sesion)):
        if sesion.iloc[i].FTR == 'H':
            teams[sesion.iloc[i].HomeTeam].append('W')
            teams[sesion.iloc[i].AwayTeam].append('L')
        elif sesion.iloc[i].FTR == 'A':
            teams[sesion.iloc[i].AwayTeam].append('W')
            teams[sesion.iloc[i].HomeTeam].append('L')
        else:
            teams[sesion.iloc[i].AwayTeam].append('D')
            teams[sesion.iloc[i].HomeTeam].append('D')
    
    tabwl = pd.DataFrame(data=teams, index = [i for i in range(1,len(list(teams.values())[0])+1)]).T
    return tabwl
```

```python
def points(sesion):
    global match_point
    
    tabwl = get_mp(sesion)
    matchres_points = tabwl.applymap(get_points)
    for i in range(2,len(list(teams.values())[0])+1):
        matchres_points[i] = matchres_points[i] + matchres_points[i-1]

    matchres_points.insert(column =0, loc = 0, value = [0*i for i in range(len(teams))])
    match_point = matchres_points
    
    sesion = sesion.sort_values(by='Id',ascending=False)
    HTP = []
    ATP = []
    j = 0
    for i in range(len(sesion)):
        ht = sesion.iloc[i].HomeTeam
        at = sesion.iloc[i].AwayTeam
        HTP.append(matchres_points.loc[ht][j])
        ATP.append(matchres_points.loc[at][j])

        if ((i + 1)% 11) == 0 and len(teams) == 22:
                j = j + 1
        elif ((i + 1)% 10) == 0 and len(teams) == 20:
            j = j + 1

    sesion['HTP'] = HTP
    sesion['ATP'] = ATP
    sesion = sesion.sort_values(by='Id',ascending=True)
    print('Completada')
    print('Temporada: ' + str(sesion.Season.unique()) + ', división: ' + sesion.Div.unique())
    return sesion
```

```python
def add_form(playing_stat,num):
    form = get_form(playing_stat,num)
    h = ['M' for i in range(num * 10)]  # since form is not available for n MW (n*10)
    a = ['M' for i in range(num * 10)]
    
    j = num
    for i in range((num*10),380):
        ht = playing_stat.iloc[i].HomeTeam
        at = playing_stat.iloc[i].AwayTeam
        
        past = form.loc[ht][j]               # get past n results
        h.append(past[num-1])                    # 0 index is most recent
        
        past = form.loc[at][j]               # get past n results.
        a.append(past[num-1])                   # 0 index is most recent
        
        if ((i + 1)% 10) == 0:
            j = j + 1

    playing_stat['HM' + str(num)] = h                 
    playing_stat['AM' + str(num)] = a

    
    return playing_stat
```

```python
def get_3w(s):
    if s[-3:] == 'WWW':
        return 1
    else:
        return 0
    
def get_5w(s):
    if s == 'WWWWW':
        return 1
    else:
        return 0
    
def get_3l(s):
    if s[-3:] == 'LLL':
        return 1
    else:
        return 0
    
def get_5l(s):
    if s == 'LLLLL':
        return 1
    else:
        return 0
```
