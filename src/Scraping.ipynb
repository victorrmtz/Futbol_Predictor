### TEAMS

```python
teams_stats = ['https://fbref.com/en/comps/12/stats/La-Liga-Stats',
               'https://fbref.com/en/comps/17/stats/Segunda-Division-Stats']
```

```python
lst_df=Parallel(n_jobs=6, verbose=True)(delayed(ex.scraping_teams)(url) for url in tqdm(teams_stats))
```

```python
len(lst_df)
```

```python
lst_df[0].columns
```

```python
lst_df[0] = lst_df[0].drop(['xG', 'npxG', 'xAG', 'npxG+xAG', 'xG', 'xAG', 'xG+xAG', 'npxG', 'npxG+xAG'], axis=1)
```

```python
lst_df[1]
```

```python
print("Saving csv")
df_teams.to_csv(r'../data/teams.csv')
print("File Saved")
```
